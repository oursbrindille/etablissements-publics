# Analyses services publics

jeu de données : https://www.data.gouv.fr/fr/datasets/annuaire-des-services-publics-nationaux/

```python
import pandas as pd
pd.set_option('display.max_colwidth', -1)
pd.set_option('display.max_rows', 1000)
df = pd.read_json('services-publics-nationaux.json', lines=True)
```

```python
df.parent_name.unique()
```

```python
df.parent_name.value_counts()
```

```python
df_filter = df[(df['parent_name'].str.contains('Gouvernement')) | (df['parent_name'].str.contains('Ministère')) | (df['parent_name'].str.contains('Premier ministre'))]
df_filter = df_filter[pd.notnull(df_filter['geo'])]
df_filter.head(3)
```

```python
df_filter.count()
```

```python
df_filter.columns
```

```python
df_filter
```

```python
df2 = df_filter
df2['address_found'] = df_filter['geo'].apply(lambda x: x.get('address_found'))
df2['address_id'] = df_filter['geo'].apply(lambda x: x.get('address_id'))
df2['address_searched'] = df_filter['geo'].apply(lambda x: x.get('address_searched'))
df2['address_type'] = df_filter['geo'].apply(lambda x: x.get('address_type'))
df2['commune'] = df_filter['geo'].apply(lambda x: x.get('commune'))
df2['insee_comm'] = df_filter['geo'].apply(lambda x: x.get('insee_comm'))
df2['latitude'] = df_filter['geo'].apply(lambda x: x.get('latitude'))
df2['longitude'] = df_filter['geo'].apply(lambda x: x.get('longitude'))
df2['score'] = df_filter['geo'].apply(lambda x: x.get('score'))
df2['source'] = df_filter['geo'].apply(lambda x: x.get('source'))
```

```python
df2.count()
```

```python
import folium
import pandas as pd
import geopandas as gpd
```

```python
m = folium.Map(location=[48.853,2.35], tiles="OpenStreetMap", zoom_start=12)
```

```python
for index, row in df2.iterrows():
    folium.Marker([row['latitude'],row['longitude']], popup='<i>'+row['name']+'<br/><br/>'+row['parent_name']+'</i>', tooltip=row['name'], icon=folium.Icon(color='red')).add_to(m)
```
